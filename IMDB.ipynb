# Top genre with highest IMDb rating

```python
df_clean = df.dropna(subset=['Genre', 'IMDB_Rating'])
df_clean['Genre'] = df_clean['Genre'].str.split(',')
df_exploded = df_clean.explode('Genre')
df_exploded['Genre'] = df_exploded['Genre'].str.strip() # To remove any whitespaces if left
genre_avg = df_exploded.groupby('Genre', as_index=False)['IMDB_Rating'].mean()
top_genres = genre_avg.sort_values(by='IMDB_Rating', ascending=False)
print(top_genres.head(10))
```

# Correlation between IMDb_rating and Runtime

```python
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')
df['IMDB_Rating'] = pd.to_numeric(df['IMDB_Rating'], errors='coerce')
```

```python
df_clean = df.dropna(subset=['Runtime', 'IMDB_Rating'])
```

```python
correlation = df_clean['Runtime'].corr(df_clean['IMDB_Rating'])
print(f"Correlation between Runtime and IMDb Rating: {correlation:.3f}")
```
